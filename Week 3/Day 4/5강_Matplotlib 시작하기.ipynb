- 가장 큰 차이점은 **제어의 역전, IoC (Inversion of Control)**

    - 프레임워크는 거꾸로 애플리케이션 코드가 프레임워크에 의해 사용됨.

### 1. Matplotlib

> 파이썬의 데이터 시각화 라이브러리

- `%matplotlib inline`을 통해서 활성화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### 2. Case Study with Arguments

```python
plt.plot([2, 4, 2, 4, 2]) # 실제 plotting을 하는 함수
plt.show() # plt를 확인하는 방법
```

#### Figsize: Figure(도면)의 크기를 설정

```python
plt.figure(figsize = (3, 3)) # plotting을 할 도면을 선언
plt.plot([1, 2, 3, 4, 5])
plt.show()
```

```python
plt.figure(figsize = (16, 16))
plt.plot([1, 2, 3, 4, 5])
plt.show()
```

#### 2차 함수 그래프 with plot()

```python
# 리스트를 이용해서 1차함수 y=x를 그려보면:

plt.plot([0, 1, 2, 3, 4])
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5]) # 정의역
y = np.array([1, 4, 9, 16, 25]) # f(x)

plt.plot(x, y)
plt.show()
```

```python
# np.arange(start, end, c) c:0.01로 매우 작게 해서 부드러운 그래프 가능

x = np.arange(-10, 10, 0.01)
plt.plot(x, x**2)
plt.show()
```

```python
# x, y축에 설명 추가하기

x = np.arange(-10, 10, 0.01)

###
plt.xlabel('x value')
plt.ylabel('y value')
###

plt.plot(x, x**2)
plt.show()
```

```python
# x, y축의 범위를 설정하기

x = np.arange(-10, 10, 0.01)

plt.xlabel('x value')
plt.ylabel('y value')

###
plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min,. y_max]
###

plt.plot(x, x**2)
plt.show()
```

```python
# x, y축에 눈금 설정하기

x = np.arange(-10, 10, 0.01)

plt.xlabel('x value')
plt.ylabel('y value')

plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min,. y_max]

###
plt.xticks([i for i in range(-5, 5, 1)]) # x 축의 눈금 설정
plt.yticks([i for i in range(0, 27, 3)]) # y 축의 눈금 설정
###

plt.plot(x, x**2)
plt.show()
```

```python
# 그래프에 title 달기

x = np.arange(-10, 10, 0.01)

plt.xlabel('x value')
plt.ylabel('y value')

plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min,. y_max]

plt.xticks([i for i in range(-5, 5, 1)]) # x 축의 눈금 설정
plt.yticks([i for i in range(0, 27, 3)]) # y 축의 눈금 설정

###
plt.title('y = x^2 graph')
###

###
plt.plot(x, x**2, label='trend')
plt.legend()
###

plt.show()
```
