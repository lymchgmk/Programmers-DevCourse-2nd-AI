# 6강. Matplotlib의 여러 plot들

## 이론

### 꺾은선 그래프 (Plot)

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.arange(20) # 0~19
y = np.random.randint(0, 20, 20) # 정수 난수를 20번 생성

plt.plot(x, y)
plt.show()
```

```python
# Extra: y축을 20까지 보이게 하고 싶다면? y축을 '5' 단위로 보이게하고 싶다면?
# .axis(), .yticks()

x = np.arange(20) # 0~19
y = np.random.randint(0, 20, 20) # 정수 난수를 20번 생성

###
plt.axis([0, 20, 0, 20])
plt.yticks([i for i in range(0, 25, 5)])
###

plt.plot(x, y)
plt.show()
```

### 산점도 (Scatter Plot)

```python
plt.scatter(x, y)
plt.show()
```

- 시계열은 꺾은선 그래프, x와 y가 상관이 없는경우 산점도를 주로 사용

### 박스 그림 (Box Plot)
> 수치형 데이터에 대한 정보 (Q1, Q2, Q3, min, max)

```python
plt.boxplot(y)

plt.show()
```

```python
# Extra: Plot의 title을 'Box plot of y'

plt.boxplot((x, y))
plt.title('Box plot of y')
plt.show()
```

### 막대 그래프 (Bar Plot)
> 범주형 데이터의 "값"과 그 값의 크기를 직사각형으로 나타낸 그림

```python
plt.bar(x, y)
plt.show()
```

```python
# Extra: xtics를 올바르게 처리해보자.

plt.bar(x, y)
plt.xticks(np.arange(0, 20, 1))

plt.show()
```

```python
# cf: Histogram
# 도수분포를 직사각형의 막대 형태로 나타낸다.
# '계급'으로 나타낸 것이 특징, 0, 1, 2가 아니라 0~2까지의 '범주형'데이터로 구성 후 그림을 그림

plt.hist(y, bins=np.arange(0, 22, 2))
```

```python
# cf: Histogram
# Extra: xticks를 올바르게 고쳐보기.


plt.hist(y, bins=np.arange(0, 22, 2))
plt.xticks(np.arange(0, 22, 2))
plt.show()
```

### 원형 그래프 (Pie Chart)
> 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프

- 다른 그래프에 비해서 비율 확인에 용이
- `.pie()`

```python
z = [100, 300, 200, 400]

plt.pie(z, labels=['one', 'two', 'three', 'four'])
plt.show()
```
