## Django

### Flask vs Django
1. **Falsk**: "***마이크로*** 웹 프레임워크" 즉 최소만 구현되어 있음. 따라서, 작은 프로젝트에 적합.
2. **Django**: "***풀 스택*** 프레임워크" 필요한 거의 대부분이 구현되어 있음. 따라서, 큰 프로젝트에 적합.

### Django의 특징

1.  MVC 패턴에 기반한 Django의 MVT 패턴
    - 용어만 다르지 개념은 동일
2. 객체 관계 매핑
    - ORM 기본 제공 
3. 자동으로 admin 페이지 생성
4. URL 디자인
    - 잘 모르겠음
5. 소스 변경사항 자동 반영
    - 소스 파일을 수정하더라도 웹서버를 다시 시작할 필요없음.
6. Instagram같은 성공적인 도입 사례가 있음.
7. 사용자 인증, 사용자 관리 등의 기능이 기본적으로 구현되어 있음.
8. Python 기반이라, 그 단점을 그대로 포함.
    - 예를 들면, 성능이 Node.js같은 다른 풀 스택 프레임워크보다 좋지 못함.

### Django의 MTV Pattern

- (일반적인) MVC 패턴: Model + View + Controller
- (Django의) MVT 패턴: Model + Template + View

#### Model (Django에서 Model)

- Controller가 호출하면, 요청에 맞는 역할을 수행하는 코드
- DB에 접근하고 CRUD를 담당

#### View (Django에서 Template)

- Controller에게 데이터를 받아 웹브라우저에 렌더링을 담당
- 사용자들이 볼 수 있는 UI를 제공하는 코드

#### Controller (Django에서 View)

- Model과 View를 이어주는 매개체를 담당
- 사용자가 Controller를 통해 Model에 데이터를 요청, 요청의 응답값을 View에 전달

#### MVC 패턴을 사용하는 이유

- 어플리케이션을 MVC 3가지로 나누어 개발에 집중가능 및 유지보수성 / 확장성 / 유연성이 증가.
- 코드의 중복성 최소화
